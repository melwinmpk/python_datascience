# Mathplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
arr1 = np.arange(0,5,0.4)
arr1
```

```python
arr2 = np.log(arr1)
arr2
```

```python
plt.plot(arr1, arr2, 'grey')  # second Parameter is the colour you can set it as 'r' => red,'b' => blue, 'g' = > green .... etc
plt.xlabel('arr1')
plt.ylabel('arr2')
plt.title('arr1 vs arr2')
```

```python
plt.subplot(1,2,1) # subplot (no of rows, no of colums, Numbers)
plt.plot(arr1,arr2,'grey')
plt.subplot(1,2,2)
plt.plot(arr2,arr1,'g')
```

```python
plt.subplot(2,2,1) # subplot (no of rows, no of colums, Numbers)
plt.plot(arr1, arr2, 'r')
plt.subplot(2,2,2)
plt.plot(arr2, arr1, 'g')
plt.subplot(2,2,3)
plt.plot(arr1, arr1*2, 'b')
plt.subplot(2,2,4)
plt.plot(arr1, arr1*3, 'grey')
```

# Object oriented methods of plotting the graph

### Its better to use  Object oriented methods of plotting the graph as we have much controll over it

```python
# Look the image for the reference
cnv = plt.figure()
ax = cnv.add_axes([0.1, 0.1, 0.8, 0.8])
# when i try to change the 'l' and 'b' it did not change it may be becoz of only one graph
# Plotting arr1 vs arr2
ax.plot(arr1, arr2, 'r', marker = 'o')
ax.set_xlabel('Arr1')
ax.set_ylabel('Arr2')
ax.set_title('Arr1 vs Arr2')
```

```python
cnv = plt.figure()
ax1 = cnv.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = cnv.add_axes([0.5, 0.2, 0.3, 0.3])
# Plotting arr1 vs arr2
ax1.plot(arr1, arr2, 'r', marker = 'o')
ax1.set_xlabel('Arr1')
ax1.set_ylabel('Arr2')
ax1.set_title('Arr1 vs Arr2')

ax2.plot(arr2, arr1, 'g', marker = 's')
ax2.set_xlabel('Arr1')
ax2.set_ylabel('Arr2')
ax2.set_title('Arr2 vs Arr1')
```

```python
cnv, ax = plt.subplots(nrows = 3, ncols = 3)
ax[0, 0].plot(arr1, arr2, 'r')
ax[0,1].plot(arr2, arr1, 'g')
plt.tight_layout()
# cnv.savefig('log_graph.png') you can save it in the Image as well even if its multiple grapg
```

```python
# you can change the size of the Graph
cnv = plt.figure(figsize=(10,5))
```

```python
# you have controle over the figure size
cnv, ax = plt.subplots(figsize=(10,5), dpi = 50) # dpi dits per Inches
ax.plot(arr1, arr2, 'r')
```

```python
cnv.savefig('log_graph.png')
```

## Setting legend

```python
cnv = plt.figure()
ax = cnv.add_axes([0.1, 0.1, 0.8, 0.8])
# Plotting arr1 vs arr2
ax.plot(arr1, arr2, 'r', marker = 'o', label = 'Arr1 vs Arr2') # you should mention the label
ax.set_xlabel('Arr1')
ax.set_ylabel('Arr2')
ax.set_title('Arr1 vs Arr2')
ax.legend() 
```

```python
cnv = plt.figure()
ax1 = cnv.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = cnv.add_axes([0.5, 0.2, 0.3, 0.3])
# Plotting arr1 vs arr2
ax1.plot(arr1, arr2, 'r--', marker = 'o', label = 'Arr1 vs Arr2',alpha = 1, lw = 0.5 )  # 'r--' i.r ('<color>--')to make the dashed grah 
#alpha Transparency of the plot just like the opacity in the css value must be 0.1 - 1 only
# lw = Line width
ax1.set_xlabel('Arr1')
ax1.set_ylabel('Arr2')
ax1.set_title('Arr1 vs Arr2')
ax1.legend()

ax2.plot(arr2, arr1, 'g--', marker = 's', label = 'Arr2 vs Arr1')
ax2.set_xlabel('Arr1')
ax2.set_ylabel('Arr2')
ax2.set_title('Arr2 vs Arr1')

ax2.legend()
```

```python
cnv = plt.figure()
ax1 = cnv.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = cnv.add_axes([0.5, 0.2, 0.3, 0.3])
# Plotting arr1 vs arr2
ax1.plot(arr1, arr2, 'r--', marker = 'o', label = 'Arr1 vs Arr2',markersize = 20, markerfacecolor = 'yellow', markeredgecolor = 'green',markeredgewidth = 2  )  # 'r--' i.r ('<color>--')to make the dashed grah 
# markersize 
ax1.set_xlabel('Arr1')
ax1.set_ylabel('Arr2')
ax1.set_title('Arr1 vs Arr2')
ax1.legend()

ax2.plot(arr2, arr1, 'g--', marker = 's', label = 'Arr2 vs Arr1')
ax2.set_xlabel('Arr1')
ax2.set_ylabel('Arr2')
ax2.set_title('Arr2 vs Arr1')

ax2.legend()
```
