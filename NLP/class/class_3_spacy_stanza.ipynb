```python
import spacy
nlp = spacy.load("en_core_web_sm")
from spacy import displacy
```

```python
text = "The Republican president is being challenged by Democratic Party nominee Joe Biden"
doc = nlp(text)

displacy.render(doc , style='dep')
```

```python
text = "The Republican president is being challenged by Democratic Party nominee Joe Biden"
doc = nlp(text)

displacy.render(doc , style='ent')
```

```python
nlp = spacy.load('en_core_web_md')

text = "The Republican president is being challenged by Democratic Party nominee Joe Biden"

doc = nlp(text)

for token in doc:
    print(token,'----->' , token.has_vector)
```

```python
nlp = spacy.load('en_core_web_md')

text = "The Republican president is being challenged by Democratic Party nominee Joe Biden Modi"

doc = nlp(text)

for token in doc:
    print(token,'----->' , token.vector)
```

```python
# similarity between the two vectors

text = 'tree' 
text_1 = 'plant' 

doc = nlp(text)
doc_1 = nlp(text_1)

doc.similarity(doc_1)
```

```python
text = 'lion bear cow apple mango spinach'

tokens = nlp(text)

for token11 in tokens:
    for token12 in tokens:
        print(token11, token12, token11.similarity(token12))
```
