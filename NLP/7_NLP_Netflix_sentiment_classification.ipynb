```python
import numpy as np
import pandas as pd
import re
import string
from nltk.corpus import stopwords
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer , CountVectorizer
from nltk.stem import WordNetLemmatizer
import spacy
```

```python
sw = stopwords.words('english')
```

```python
# read the psoitive data

pos_rev = pd.read_csv('netflix/pos.txt', sep='&&',header = None , encoding='latin-1')
pos_rev['mood'] = 1
pos_rev.rename(columns = {0:'review'}, inplace=True)
pos_rev.head(5)
```

```python
# read the negative data

neg_rev = pd.read_csv('netflix/negative.txt' , sep = '&&' , header = None , encoding='latin-1')
neg_rev['mood'] = 0
neg_rev.rename(columns = {0:'review'}, inplace=True)
neg_rev.head(5)
```

```python
com_rev = pd.concat([neg_rev , pos_rev] , axis = 0 , ignore_index = True )
```

```python
#1 .lower
#2. spaces
#3. punctatution
#4. stopwrods
#5. lemmatize
```

```python
string.punctuation
```

```python
com_rev['review'] = com_rev['review'].apply(lambda x : x.lower())
com_rev['review'] = com_rev['review'].apply(lambda x : re.sub(r"[^a-zA-Z]", " " , x))
com_rev['review'] = com_rev['review'].apply(lambda x : " ".join([word for word in x.split() if (word not in string.punctuation) and (word not in (sw))]))
com_rev.head()
```

```python
com_rev
```

```python
# Lemmatizer
nlp = spacy.load('en_core_web_sm', disable=['parser', 'ner'])
def space(comment):
    doc = nlp(comment)
    return " ".join([token.lemma_ for token in doc])

com_rev['review'] = com_rev['review'].apply(space)
```

```python
# train test split

X_train , X_test , y_train , y_test = train_test_split(com_rev['review'].values , com_rev['mood'].values ,
                                                      test_size = 0.2 , random_state = 101)
```

```python
train_data = pd.DataFrame({'review':X_train , 'mood':y_train})
test_data = pd.DataFrame({'review':X_test , 'mood':y_test})
```

```python
train_data.head()
```

```python
test_data.head()
```

```python
vectorizer = TfidfVectorizer()
train_vectores = vectorizer.fit_transform(train_data.review)
test_vectores = vectorizer.transform(test_data.review)
```

```python
len(vectorizer.get_feature_names())
```

```python
vectorizer.get_feature_names()
```

# ML model

```python
from sklearn import svm
from sklearn.metrics import classification_report
```

```python
classifer = svm.SVC()
classifer.fit(train_vectores , train_data.mood)
```

```python
pred = classifer.predict(test_vectores)
```

```python
report = classification_report(test_data.mood , pred , output_dict=True)
report
```
