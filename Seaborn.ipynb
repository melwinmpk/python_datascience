# Seaborn

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pok_data = pd.read_csv('dataset/Pokemon.csv')
```

```python
pok_data.head()
```

```python
pok_data.describe()
```

```python
# sns.distplot() will give histogram and kde curve both for a univariate plot
grph = sns.distplot(pok_data['Health Point'], color = 'Green')
grph.set_title('Health Point vs Frequency')
grph.set_xlabel('Health Point')
grph.set_ylabel('Frequency')
```

```python
sns.distplot(pok_data['Health Point'], kde = False)
```

```python
grph = sns.distplot(pok_data['Health Point'], color = 'red', hist = False)
grph.set_title('Health Point vs Frequency')
grph.set_xlabel('Health Point')
grph.set_ylabel('Frequency')
```

```python
# rug plot inside distplot
grph = sns.distplot(pok_data['Health Point'], rug = True)
grph.set_title('Health Point vs Frequency')
grph.set_xlabel('Health Point')
grph.set_ylabel('Frequency')
```

```python
sns.rugplot(pok_data['Health Point'], axis = 'x', height = 0.5)
```

```python
sns.distplot(pok_data['Health Point'], kde = False, rug = True, bins = 100)
```

```python
sns.distplot(pok_data['Health Point'], hist = False, rug = True, bins = 100)
```

```python
sns.set(color_codes = True)
sns.kdeplot(pok_data['Health Point'], shade = True, color = 'green')
sns.rugplot(pok_data['Health Point'], axis = 'x')
```

```python
# plotting bivariate data
sns.jointplot(x = 'Health Point', y = 'Total Point', data = pok_data)
```

```python
sns.jointplot(x = 'Health Point', y = 'Total Point', data = pok_data, xlim= {0,500}, ylim = {0,1000})
```
