# Deaths and recoveries per 100 cases

```python
daywise.head()
```

```python
fig_c = px.bar(daywise,
               x = 'Date',
               y = 'Confirmed',
               color_discrete_sequence = ['Orange'])
fig_d = px.bar(daywise,
               x = 'Date',
               y = 'Confirmed',
               color_discrete_sequence = ['Blue'])
fig_d_100 = px.bar(daywise,
               x = 'Date',
               y = 'Deaths / 100 Cases',
               color_discrete_sequence = ['Red'])
fig_r_100 = px.bar(daywise,
               x = 'Date',
               y = 'Recovered / 100 Cases',
               color_discrete_sequence = ['Green'])

fig = make_subplots(rows = 2, cols = 2,
                    shared_xaxes = False,
                    horizontal_spacing = 0.1, # 0.9
                    vertical_spacing = 0.2,
                    subplot_titles=('Confirmed Cases', 'Death Cases', 'Deaths / 100 Cases','Recovered / 100 Cases' ))
fig.add_trace(fig_c['data'][0],row = 1,col=1)
fig.add_trace(fig_d['data'][0],row = 1,col=2)
fig.add_trace(fig_d_100['data'][0],row = 2,col=1)
fig.add_trace(fig_r_100['data'][0],row = 2,col=2)

fig.update_layout(height = 600)
fig.show()
```

# No of cases daywise & No of Countries affected daywise

```python
fig_ca_d = px.bar(daywise,
               x = 'Date',
               y = 'Confirmed',
               color_discrete_sequence = ['Blue'])
fig_co_d = px.bar(daywise,
               x = 'Date',
               y = 'No. of Countries',
               color_discrete_sequence = ['Red'])

fig = make_subplots(rows = 1, cols = 2,
                    shared_xaxes = False,
                    horizontal_spacing = 0.1, # 0.9
                    vertical_spacing = 0.2,
                    subplot_titles=('Daywise Cases', 'Countries Affected DayWise'))
fig.add_trace(fig_ca_d['data'][0],row = 1,col=1)
fig.add_trace(fig_co_d['data'][0],row = 1,col=2)
fig.update_layout(height = 600)
fig.show()
```
